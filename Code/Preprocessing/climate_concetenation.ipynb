```python
import os
import pandas as pd
```

```python
#now mount to drive and set the working directory to our shared drive
from google.colab import drive
drive.mount('/content/drive')
```

## Concatenate historical and export

```python
scenario = 'historical'
scenario_path = '/content/drive/Shareddrives/Spatial_Term_Paper/data/malawi/climate_csv/' + scenario + '/'
os.chdir(scenario_path)
variables = list(dict.fromkeys(os.listdir(), None))
variables.remove('.DS_Store')
variables
```

```python
for variable in variables:
  variable_path = scenario_path + '/' + variable + '/'
  os.chdir(variable_path)
  files = list(dict.fromkeys(os.listdir(), None))
  print(len(files))
```

```python
scenario_concat = pd.DataFrame(columns = ['time', 'lat', 'lon', 'value', 'variable', 'scenario'])
for variable in variables:
  variable_path = scenario_path + '/' + variable + '/'
  os.chdir(variable_path)
  files = list(dict.fromkeys(os.listdir(), None))
  for file in files:
    df = pd.read_csv(file, index_col=0)
    df = df.rename(columns = {variable:'value'})
    scenario_concat = pd.concat([scenario_concat, df], ignore_index=False)
```

```python
scenario_concat
```

```python
saving_path = '/content/drive/Shareddrives/Spatial_Term_Paper/data/concatenated/climate/' + scenario + '/'
os.chdir(saving_path)
```

```python
scenario_concat.to_csv(scenario + '_all_variables.csv')
```
