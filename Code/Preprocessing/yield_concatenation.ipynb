```python
import numpy as np
import pandas as pd
import os
```

```python
#now mount to drive and set the working directory to our shared drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/Shareddrives/Spatial_Term_Paper/data/yields/crop_yield_data_csv/'
os.chdir(path)
directories = dict.fromkeys(os.listdir(), None)
directories
```

```python
seasons = ['maize', 'maize_major', 'maize_second']
```

```python
yields_concat = pd.DataFrame(columns = ['latitude', 'longitude', 'crop_yield', 'year', 'crop_type'])
for season in seasons:
  navigation = path + '/' + season
  os.chdir(navigation)
  directories = dict.fromkeys(os.listdir(), None)
  for file in directories.keys():
    #because the zip files are in there we need the if statement
    if file.endswith(".csv"):
      yields_concat = pd.concat([yields_concat, pd.read_csv(file, index_col=0)], ignore_index=False)
      
```

```python
all_seasons = yields_concat.reset_index(drop=True)
```

```python
saving_path = '/content/drive/Shareddrives/Spatial_Term_Paper/data/concatenated/yields'
os.chdir(saving_path)
```

```python
all_seasons[all_seasons['crop_type'] == 'maize'].to_csv('maize.csv')
all_seasons[all_seasons['crop_type'] == 'maize_major'].to_csv('maize_major.csv')
all_seasons[all_seasons['crop_type'] == 'maize_second'].to_csv('maize_second.csv')
```
